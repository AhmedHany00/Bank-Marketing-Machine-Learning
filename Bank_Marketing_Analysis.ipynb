# Task 1: Dataset Description

**Objective:** Analyze the `bank-additional-full.csv` dataset, describing feature types, missing values, outliers, dimensionality, and the target class.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Note: The dataset often uses semicolons as separators. Adjusting accordingly.
file_path = 'bank-additional-full.csv'
try:
    df = pd.read_csv(file_path, sep=';')
    print("Dataset loaded successfully with separator ';'.")
except:
    df = pd.read_csv(file_path, sep=',')
    print("Dataset loaded successfully with separator ','.")

# Display the first few rows
display(df.head())
```

## 1. Dimensionality

```python
print(f"Dimensionality of the dataset: {df.shape}")
print(f"Number of rows: {df.shape[0]}")
print(f"Number of columns: {df.shape[1]}")
```

## 2. Feature Types

```python
df.info()
```

```python
# Separate numerical and categorical features
numerical_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(exclude=[np.number]).columns.tolist()

print("Numerical Features:", numerical_cols)
print("Categorical Features:", categorical_cols)
```

## 3. Missing Values

```python
# Check for standard missing values (NaN)
missing_values = df.isnull().sum()
missing_percentage = (missing_values / len(df)) * 100

print("Missing Values per Column:")
print(missing_values[missing_values > 0])

# Check for potential 'unknown' values which are common in this dataset
print("\nOccurrences of 'unknown' in categorical columns:")
for col in categorical_cols:
    unknown_count = (df[col] == 'unknown').sum()
    if unknown_count > 0:
        print(f"{col}: {unknown_count} ({unknown_count/len(df)*100:.2f}%)")
```

## 4. Outliers (Numerical Features)

```python
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 4, i + 1)
    sns.boxplot(y=df[col])
    plt.title(col)
plt.tight_layout()
plt.show()

# Calculate IQR based outliers
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"{col}: {len(outliers)} outliers ({len(outliers)/len(df)*100:.2f}%)")
```

## 5. Target Class Distribution

```python
target_col = 'y'  # Usual target for bank dataset
if target_col in df.columns:
    print(df[target_col].value_counts())
    print(df[target_col].value_counts(normalize=True) * 100)
    sns.countplot(x=target_col, data=df)
    plt.title('Target Class Distribution')
    plt.show()
else:
    print(f"Target column '{target_col}' not found. Please check column names.")
```
